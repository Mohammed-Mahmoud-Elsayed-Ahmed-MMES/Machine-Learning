```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('hour.csv')
dataset.head(5)
```

```python
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X.shape)
print(X)
```

```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# Assuming 'dataset' is your DataFrame containing the features

try:
  # Convert the date column to datetime format (assuming it's parsable)
  dataset['dteday'] = pd.to_datetime(dataset.iloc[:, 1])  # Assign to existing column

  # Create a DataFrame with only the date column
  dates_df = dataset[['dteday']]

  # Apply one-hot encoding to the date DataFrame
  ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(sparse=False), [0])])
  encoded_dates = ct.fit_transform(dates_df)

  # Drop the original date column and add the encoded columns to dataset
  dataset.drop('dteday', axis=1, inplace=True)
  for col in range(encoded_dates.shape[1]):  # Assuming columns from encoding
    dataset['date_encoded_' + str(col+1)] = encoded_dates[:, col]

except (ValueError, AttributeError):  # Handle cases where date format is invalid
  print("Error: Could not parse date column. Please check the data format.")
```
